LIMPIEZA COLUMNA CASENUMBER

```python
#casenumber es un ID por lo que cada caso es único, por lo que sustituiremos todos los valores de 1 -n

n = sharks.shape[0]

sharks['casenumber'] = sharks['casenumber'].apply(lambda x: next(n for n, val in zip(range(0, n+1), sharks['casenumber']) if val == x))

sharks.head()
```

LIMPIEZA COLUMNA DATE

```python
#método para limpiar la columna date:
#como en todos los valores de casenumber.1 aparece la fecha, crearemos una nueva columna que se llame date_ok (dejando solo la fecha de casenumber.1 )
```

```python
sharks['date_ok']= sharks['casenumber.1']

sharks['date_ok']= sharks['date_ok'].str.extract(r'([0-9]{4}.[0-9]{2}.[0-9]{2})', expand=False)

#si no consigue dejar sólo lo que sería una fecha, devuelve Nan. Por ello, en los que no tenemos clara la fecha del ataque, borramos las filas

sharks.date_ok.isna().sum()

sharks = sharks.dropna(subset=['date_ok'])

sharks.date_ok.value_counts().head()

#Como podemos ver, aunque todas los registros de nuestra columna siguen la misma estructura de fecha 'YY.MM.DD', en algunos casos el mes y el día aparecen como 0, por lo que también procederemos a limpiar esas fechas al considerarlas como inválidas.

sharks = sharks[~sharks['date_ok'].astype(str).str.contains('.00')]
sharks.date_ok.value_counts().head()

#Con ello, nos hemos quedado únicamente con los registros que contienen fechas válidas


```

```python
sharks.shape
```

LIMPIEZA DE LA COLUMNA AGE

```python
sharks.age.unique()
```

```python
#se trata de una columna que además de contener 1475 valores Nan, presenta edades representadas con integers, caracteres especiales o palabras y frases.

#Procedimiento de limpieza, relleno de nulos:
  # Aunque la variable edad no sea significativa para nuestro análisis, para una mayor consistencia de los datos, rellenaremos los nulos con 'boat' cuando el ataque haya sido a un barco y con 'unknown' en el resto de casos:

    
sharks.loc[sharks['age'].isnull(), 'type'].unique()

sharks.loc[sharks['type'] == 'Provoked', 'age'] = sharks.loc[sharks['type'] == 'Provoked', 'age'].fillna('unknown')

sharks.loc[sharks['type'] == 'Unprovoked', 'age'] = sharks.loc[sharks['type'] == 'Unprovoked', 'age'].fillna('unknown')

sharks.loc[sharks['type'] == 'Boating', 'age'] = sharks.loc[sharks['type'] == 'Boating', 'age'].fillna('boat')

sharks.age.value_counts().head()   
 
 
```

```python
   
# Unificar todos los valores 'extraños':
 #factores a tener en cuenta:
    # todavía nos podemos encontrar casos en los que la edad no tenga sentido porque fue un ataque a un barco, por lo que reemplazaremos los datos por boat cuando se de el caso.
        
sharks.loc[sharks['type'] == 'Boating', 'age'] = 'boat'
     
        
    # En algunas ocasiones aparecen dos edades de la siguiente forma '17 & 16'. En estos casos, entendemos que fueron atacadas más de una persona por las que las dejaremos así.
    
    #Hay casos que hay que tratar aisladamente, para lo que usaremos la función replace:
    
sharks = sharks.replace({'age': {'teen':'unknown','20?':'unknown',' ':'unknown','21 or 26':'unknown','18 to 22':'unknown',' 43':'43','>50':'unknown','adult':'unknown','(adult)':'unknown','25 or 28':'unknown','13 or 18': 'unknown','Both 11':'11', '  ': 'unknown','young': 'unknown','9 or 10': 'unknown', '2½': '2', '18 months': '2', 'A.M.': 'unknown', 'Teen':'unknown','"young"':'unknown', 'Elderly':'unknown','31 or 33':'unknown', '10 or 12':'unknown','X':'unknown'}})
    
    #En muchas ocasiones la edad va seguida de una s, eliminaremos la letra:
sharks['age'] = sharks['age'].str.replace('s', '')
```

LIMPIEZA DE LA COLUMNA SPECIES

```python
#Limpieza de valores nulos. Como es una columna categórica y no es relevante el tipo de tiburón para el análisis. Rellenaremos con 'unknown'

sharks.fillna({'species':'unknown'}, inplace=True)
```

```python
sharks.species.unique()
```

```python
# En la mayoría de casos, antes de la palabra shark aparece o bien la raza o bien el tamaño, ambos nos sirven para dejar un poco más limpia la columna
# Crear función para limpiar valores de species
def clean_species(x):
    if 'shark' in x:
        words = x.split()
        if 'shark' in words:  
            index = words.index('shark')
            return words[index-1] + ' ' + 'shark'
    else:
        return 'unknown'

sharks['species'] = sharks['species'].apply(clean_species)



sharks.species.value_counts().head()


#para unificar un poco más los valores, pasaremos todo a minúsculas y quitaremos algunos caracteres especiales:
```

```python
sharks.fillna({'species':'unknown'}, inplace=True) #aunque antes habíamos rellenado los nulos, parece que la función creada puede devolver Nan, por lo que repetimos el proceso quí también.
```

```python
sharks['species'] = sharks['species'].str.replace('[', '').str.replace(']', '').str.replace('', '').str.replace('', '').str.lower()

sharks.species.value_counts().head()

```
