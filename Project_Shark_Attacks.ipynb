```python
## Exploración de columnas.

#Para la exploración de columnas primero vamos a 'unificar los nombres de las columnas' para evitar problemas de caracteres especiales y case sensitive.

```

```python
sharks.columns = [e.lower(). replace(' ', '').replace(':', '').replace('(','').replace(')','').replace('/', '') for e in sharks.columns]

sharks.head()
```

```python
plt.figure(figsize=(10, 6))  

sns.heatmap(sharks.isna(),   
            yticklabels=False,
            cmap='viridis',
            cbar=False
           )

plt.show();
```

```python
#Ahora vamos a eliminar las filas duplicadas , primero comprobando si hay filas duplicadas:
sharks.duplicated().any()
```

```python
sharks.drop_duplicates(inplace=True)

sharks.shape

#hemos eliminado 2392 filas que no aportaban valor
```

```python
plt.figure(figsize=(10, 6))  

sns.heatmap(sharks.isna(),   
            yticklabels=False,
            cmap='viridis',
            cbar=False
           )

plt.show();
```

```python
#Tal y como se ve en la gráfica, ahora no todas las columnas tienen nulos (o algunas tienen un número relativamente pequeño respecto a la muestra). Por ello, vamos a explorar que columnas son y tratar de identificar la mejor solución para cada una. 

nan_cols2 = sharks.isna().sum()

nan_cols2[nan_cols2>0].sort_values(ascending=True)
```

```python
#la media de valores nulos por columna es de 1032, pero hay outliers por lo que el parámetro de guía para seguir explorando debería ser la mediana.
nan_cols2[nan_cols2>0].sort_values(ascending=True).mean()
```

```python
nan_cols2[nan_cols2>0].sort_values(ascending=True).median()
```

```python
#Borrado de filas: como aún tenemos una muestra relativamente grande,vamos a ver la distribución de Nan por filas para tomar una decisión. 
#crearemos una variable 'nan_filas' y añadiremos una columana nueva con el número total de nans que hay en cada fila. 

nan_filas = sharks.isna().sum(axis=1)

sharks['nan_filas'] = nan_filas

sharks.head()




```

```python
sharks.shape
```

```python
sharks.nan_filas.max(),sharks.nan_filas.min()
```

```python
sharks.nan_filas.unique(), sharks.nan_filas.median() #como hay outliers, tomamos la mediana como parámetro para el borrado de filas
```

```python
#borramos las filas que tengan dos o menos nan

sharks = sharks.dropna(thresh=4, axis=0)

sharks.shape

#hemos borrado todas las filas que tenían sólo un valor no nulo (25-1 columnas). En total 3393.
```

```python
sharks.head()
```

```python
nan_cols3 = sharks.isna().sum()

nan_cols3[nan_cols3>0].sort_values(ascending=True)
```

```python
plt.figure(figsize=(10, 6))  

sns.heatmap(sharks.isna(),   
            yticklabels=False,
            cmap='viridis',
            cbar=False
           )

plt.show();
```

```python
#limpieza de las columnas unnamed22 y unnamed23, que al ser columnas con casi todos los valores NaN no aportarán valor al análisis y por tanto no nos interesan. Por otro lado, como son columnas categóricas,  nos limitaremos a rellenar el NaN con 'Unknown')
sharks.fillna({'unnamed22':'unknown','unnamed23':'unknown'}, inplace=True)
```
